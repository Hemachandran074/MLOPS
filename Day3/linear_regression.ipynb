```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
```

```python
data.head()
```

```python
x = data.drop('per capita income (US$)', axis='columns')
y = data['per capita income (US$)']
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
pred = model.predict([[2030]])
```

```python
print(pred[0])
```

```python
plt.scatter(data['year'], data['per capita income (US$)'])
plt.plot(data['year'], model.predict(x))

plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")

plt.show()
```
